## Use Geographic Information Systems to perform spatial data analysis and make maps

* A system that allows you to visualize, analyze, and store data to understand spatial relationships, patterns, and trends

* GIS is widely used in organizations of all sizes and in almost every industry

* A mix of data, science, analysis, and maps

## We Can Answer _Spatial_ Questions

* Where are there hot spots of crime in a city?

* Where should we build a store/wind turbine/solar panel/ski resort?

* Where will a natural disaster hit and what cities will be impacted?

* Where is the highest concentration of arsenic in our port?

* How much more nitrogen will end up in a watershed if we turn a forested area into a commercial area?

```python
import geopandas as gpd
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(figsize=(15,10))
ax.set_title("WGS84 (Latitude/Longitude)")
ax.grid('on', which='major', axis='x', linewidth=0.5)
ax.grid('on', which='major', axis='y', linewidth=0.5)
```

![title](img/epsgIO.png)

Source: [epsg.io](http://epsg.io/)

```python
world_sp = world.to_crs(epsg=3031)
ax = world_sp.plot(figsize=(15,10))
ax.set_title("Antarctic Polar Stereographic")
ax.grid('on', which='major', axis='x', linewidth=0.5)
ax.grid('on', which='major', axis='y', linewidth=0.5)
```

```python
world_mercator = world.to_crs(epsg=3395)
ax = world_mercator.plot(figsize=(15,10))
ax.set_title("Mercator")

ax.grid('on', which='major', axis='x', linewidth=0.5)
ax.grid('on', which='major', axis='y', linewidth=0.5)
```

```python
world_mercator = world_mercator[(world_mercator.name != "Antarctica") & (world_mercator.name != "Fr. S. Antarctic Lands")]
ax = world_mercator.plot(figsize=(15,10))
ax.set_title("Mercator not Including Antarctica")

ax.grid('on', which='major', axis='x', linewidth=0.5)
ax.grid('on', which='major', axis='y', linewidth=0.5)
```
